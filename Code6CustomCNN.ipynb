```python
import os
import cv2
import numpy as np
import tensorflow as tf
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from google.colab import drive
import matplotlib.pyplot as plt
import seaborn as sns
import random

# Mount Google Drive
drive.mount('/content/drive')

# Set dataset path
DATASET_PATH = '/content/drive/MyDrive/modifdataset'  # Replace with your dataset folder path

# 1. Load Dataset
def load_dataset(dataset_path):
    data = []
    labels = []
    classes = sorted(os.listdir(dataset_path))  # Sorted for consistent class mapping
    for class_idx, class_dir in enumerate(classes):
        class_path = os.path.join(dataset_path, class_dir)
        if os.path.isdir(class_path):
            for img_file in os.listdir(class_path):
                img_path = os.path.join(class_path, img_file)
                img = cv2.imread(img_path)
                if img is not None:
                    img = cv2.resize(img, (224, 224))  # Resize to standard size
                    data.append(img)
                    labels.append(class_idx)
    return np.array(data), np.array(labels), classes

print("Loading dataset...")
images, labels, class_names = load_dataset(DATASET_PATH)
print(f"Dataset loaded: {len(images)} images across {len(class_names)} classes.")

# 2. Hybrid Feature Extraction (Gaussian + Rule-Based)
def extract_skin_mask(image):
    # Convert to HSV
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Rule-Based Mask
    lower_hsv = np.array([0, 48, 80], dtype="uint8")
    upper_hsv = np.array([20, 255, 255], dtype="uint8")
    rule_based_mask = cv2.inRange(hsv_image, lower_hsv, upper_hsv)

    # Gaussian Blur on the mask
    gaussian_mask = cv2.GaussianBlur(rule_based_mask, (5, 5), 0)

    # Combine masks
    combined_mask = cv2.bitwise_and(rule_based_mask, gaussian_mask)
    return combined_mask

print("Extracting skin masks...")
skin_masks = np.array([extract_skin_mask(img) for img in images])

# Normalize images and skin masks
images = images / 255.0
skin_masks = skin_masks / 255.0

# Add skin mask as a 4th channel to the images
print("Combining RGB channels with skin masks...")
images_with_mask = np.concatenate((images, np.expand_dims(skin_masks, axis=-1)), axis=-1)

# 3. Visualize Preprocessed Images
def visualize_preprocessed_images(images, skin_masks, labels, class_names, num_samples=5):
    plt.figure(figsize=(15, len(class_names) * 4))
    for class_idx, class_name in enumerate(class_names):
        class_indices = np.where(labels == class_idx)[0]
        selected_indices = random.sample(list(class_indices), min(num_samples, len(class_indices)))
        for i, idx in enumerate(selected_indices):
            plt.subplot(len(class_names), num_samples * 2, class_idx * num_samples * 2 + i * 2 + 1)
            plt.imshow(images[idx])
            plt.title(f"{class_name} (RGB)")
            plt.axis('off')
            plt.subplot(len(class_names), num_samples * 2, class_idx * num_samples * 2 + i * 2 + 2)
            plt.imshow(skin_masks[idx], cmap='gray')
            plt.title(f"{class_name} (Skin Mask)")
            plt.axis('off')
    plt.tight_layout()
    plt.show()

print("Visualizing preprocessed images...")
visualize_preprocessed_images(images, skin_masks, labels, class_names, num_samples=5)

# 4. Split Dataset
print("Splitting dataset...")
X_train, X_test, y_train, y_test = train_test_split(images_with_mask, labels, test_size=0.3, random_state=42)
y_train = to_categorical(y_train, num_classes=len(class_names))
y_test = to_categorical(y_test, num_classes=len(class_names))
print(f"Training samples: {len(X_train)}, Testing samples: {len(X_test)}")

# 5. Build Custom CNN Model
def create_custom_cnn(input_shape, num_classes):
    model = Sequential([
        # First Convolutional Block
        Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=input_shape),
        MaxPooling2D(pool_size=(2, 2)),
        Dropout(0.25),

        # Second Convolutional Block
        Conv2D(64, (3, 3), activation='relu', padding='same'),
        MaxPooling2D(pool_size=(2, 2)),
        Dropout(0.25),

        # Third Convolutional Block
        Conv2D(128, (3, 3), activation='relu', padding='same'),
        MaxPooling2D(pool_size=(2, 2)),
        Dropout(0.25),

        # Flatten and Fully Connected Layers
        Flatten(),
        Dense(256, activation='relu'),
        Dropout(0.5),
        Dense(num_classes, activation='softmax')
    ])
    # Hyperparameters
    learning_rate = 0.001  # Define your learning rate here

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
        #optimizer='adam', learning_rate = learning_rate,
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

print("Building custom CNN model...")
input_shape = (224, 224, 4)  # Input shape includes the skin mask
num_classes = len(class_names)
custom_cnn_model = create_custom_cnn(input_shape, num_classes)

# Display model summary
custom_cnn_model.summary()



# 6. Train the Model
print("Training the custom CNN model...")
history = custom_cnn_model.fit(X_train, y_train, epochs=100, batch_size=64, validation_split=0.3)

# 7. Evaluate the Model
print("Evaluating the model...")
test_loss, test_accuracy = custom_cnn_model.evaluate(X_test, y_test, verbose=0)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")

# Confusion Matrix
print("Confusion Matrix:")
y_pred = np.argmax(custom_cnn_model.predict(X_test), axis=1)
y_true = np.argmax(y_test, axis=1)
conf_matrix = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

# Classification Report
print("Classification Report:")
print(classification_report(y_true, y_pred, target_names=class_names))

# Visualize a random test image with prediction
def visualize_test_prediction(X_test, y_true, y_pred, class_names, idx=None):
    """
    Visualize a single test image along with its predicted and true label.
    """
    if idx is None:
        idx = random.randint(0, len(X_test) - 1)  # Choose a random index if none provided

    # Get the test image, true label, and predicted label
    test_image = X_test[idx]
    true_label = class_names[y_true[idx]]
    predicted_label = class_names[y_pred[idx]]

    # Rescale the image for display
    test_image_rgb = test_image[:, :, :3]  # Exclude skin mask for display
    test_image_rgb = (test_image_rgb * 255).astype('uint8')  # Convert back to original range

    # Display the image and labels
    plt.figure(figsize=(6, 6))
    plt.imshow(test_image_rgb)
    plt.title(f"True: {true_label}\nPredicted: {predicted_label}")
    plt.axis('off')
    plt.show()

# Display a random test image with prediction
visualize_test_prediction(X_test, y_true, y_pred, class_names)

# 8. Plot Training History
def plot_training_history(history):
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Accuracy Over Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Loss Over Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.tight_layout()
    plt.show()

plot_training_history(history)
```
